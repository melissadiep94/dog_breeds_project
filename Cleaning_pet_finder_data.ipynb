```python
# Module for reading CSV files
import csv

csvpath = os.path.join('Resources', 'original', 'df_breed_labels_api.csv')

#define the lists
rows = []
fields = []
breed_IDs = []
# Method 2: Improved Reading using CSV module

with open(csvpath, newline= '') as csvfile:

    # CSV reader specifies delimiter and variable that holds contents
    csvreader = csv.reader(csvfile, delimiter=',')

    #print(csvreader)

    # Read the header row first (skip this step if there is no header)
    fields = next(csvreader)
    #print(f"Field Names: {fields}")

    #calculating # of rows and total revenue
    
    for row in csvreader:
        rows.append(row)

    
    i = 0 
    for i in range(len(rows)):
        # Take the difference between two months 
        breed_ID = i + 1
        # append the monthly profit/loss calculation to pl_change list
        breed_IDs.append(breed_ID)

print(breed_IDs)
```

```python
dog_breeds_df["breed_ID"] = breed_IDs
#Rename columns
dog_breeds_df= dog_breeds_df.rename(columns = {'name' : 'breed_name', 'breed': 'animal_type'})
#Reorder columns
dog_breeds_df = dog_breeds_df.reindex(columns= ['breed_ID', 'breed_name', 'animal_type'])

dog_breeds_df
```
