```python
import os
import pandas as pd
```

```python
breed_weight_path = os.path.join("Resources","original","breed_weight.csv")

df_breed_weight = pd.read_csv(breed_weight_path) 
df_breed_weight
```

```python
#Drop original name
df_breed_weight= df_breed_weight.drop(columns=['Breed'])
```

```python
#Rename column headers
df_breed_weight= df_breed_weight.rename(columns = {'Breed_clean' : 'breed_name', 'BreedID': 'breed_ID', 'Weight Male': 'weight_male', 'Weight Female': 'weight_female'})
```

```python
#Drop N/As
df_breed_weight= df_breed_weight[df_breed_weight!=0].dropna()
```

```python
#Reorder columns
df_breed_weight = df_breed_weight.reindex(columns= ['breed_ID', 'breed_name', 'weight_male','weight_female'])
df_breed_weight
```

```python
#Output pandas dataframe into CSV file and store it in 'clean' folder under Resources folder
output_path = os.path.join("Resources","clean","df_breed_weight_clean.csv")
print(output_path)
df_breed_weight.to_csv(output_path)
```
