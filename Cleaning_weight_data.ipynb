```python
import os
import pandas as pd
import numpy
```

```python
breed_weight_path = os.path.join("Resources","original","breed_weight.csv")

df_breed_weight = pd.read_csv(breed_weight_path) 
df_breed_weight
```

```python
#Drop original name
df_breed_weight_v2 = df_breed_weight.drop(columns=['Breed'])
df_breed_weight_v2
```

```python
# Optional: Get all the data for the duplicate dog breeds

df_duplicate = df_breed_weight_v2[df_breed_weight_v2['Breed_clean'] == "Jack Russell Terrier"]
df_duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

df_breed_weight_v2 = df_breed_weight_v2.drop(index=96)
df_breed_weight_v2
```

```python
#Rename column headers
df_breed_weight_v3= df_breed_weight_v2.rename(columns = {'Breed_clean' : 'breed_name', 'BreedID': 'breed_IDv1', 'Weight Male': 'weight_male', 'Weight Female': 'weight_female'})
df_breed_weight_v3
```

```python
dogs_list_path = os.path.join("Resources","clean","df_breed_labels_api_clean.csv")

df_breed_labels = pd.read_csv(dogs_list_path) 
#Drop unnamed 0
df_breed_labels= df_breed_labels.drop(columns=['Unnamed: 0'])
df_breed_labels
```

```python
breed_ID_mapping = dict(zip(df_breed_labels.breed_name, df_breed_labels.breed_ID))
```

```python
df_breed_weight_v3['breed_ID'] = df_breed_weight_v3.breed_name.map(breed_ID_mapping)


df_breed_weight_v3
```

```python
df_breed_weight_v3= df_breed_weight_v3.drop(columns=['breed_IDv1'])
df_breed_weight_v3
```

```python
#Reorder columns
df_breed_weight_v3 = df_breed_weight_v3.reindex(columns= ['breed_ID', 'breed_name', 'weight_male','weight_female'])
```

```python
#Drop N/As
df_breed_weight_v3= df_breed_weight_v3[df_breed_weight_v3!=0].dropna()
df_breed_weight_v3 = df_breed_weight_v3.sort_values(by=['breed_ID'])
df_breed_weight_v3
```

```python
#Output pandas dataframe into CSV file and store it in 'clean' folder under Resources folder
output_path = os.path.join("Resources","clean","df_breed_weight_clean_v2.csv")
print(output_path)
df_breed_weight_v3.to_csv(output_path)
```
