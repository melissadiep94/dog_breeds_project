```python
import pandas as pd
from sqlalchemy import create_engine
import json
import os
```

```python
filepath = os.path.join("Resources", "dogs.json")
with open(filepath) as jsonfile:
    video_json = json.load(jsonfile)
```

```python
df = pd.DataFrame(video_json['breeds'])
```

```python
df
df.rename(columns={'name':'breed_name', 
                   'kidFriendly': 'kid_friendly', 
                   'dogFriendly': 'dog_friendy', 
                   'lowShedding': 'low_shedding',
                   'easyToGroom': 'easy_to_groom', 
                   'highEnergy': 'high_energy', 
                   'goodHealth': 'good_health', 
                   'lowBarking': 'low_barking', 
                   'easyToTrain': 'easy_to_train', 
                   'toleratesHot': 'tolerates_hot',
                   'toleratesCold': 'tolerates_cold'}, inplace=True)

pd.set_option('precision', 0)
```
