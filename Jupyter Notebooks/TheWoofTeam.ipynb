```python
import sqlalchemy 
import pandas as pd 
import numpy
from sqlalchemy import create_engine
```

```python
sql_connection_string = "postgres:postgres@localhost:5432/TheWoofTeam"
engine = create_engine(f"postgresql://{sql_connection_string}")
```

```python
engine.table_names()
```

```python
dog_api_df = pd.read_csv("df_petfinder_dogslist_api_clean_v2.csv")
print(dog_api_df.head())
```

```python
new_dog_api_df = dog_api_df.rename(columns=
                                   {"dog_ID":"dog_id",
                                    "dog_name":"dog_name", 
                                    "url_petfinder":"url_petfinder", 
                                    "dog_age":"dog_age", 
                                    "dog_gender":"dog_gender",
                                    "dog_size":"dog_size", 
                                    "dog_coat":"dog_coat", 
                                    "dog_description":"dog_decsription", 
                                    "published_at":"published_at",
                                    "distance_miles_from_07090":"distance_miles_from_07090", 
                                    "breed_ID":"breed_id", 
                                    "breed":"breed_name", 
                                    "colors.primary":"colors_primary", 
                                    "attributes.spayed_neutered":"attributes_spayed_neutered", 
                                    "attributes.house_trained":"attributes_house_trained", 
                                    "attributes.special_needs":"attributes_special_needs", 
                                    "attributes.shots_current":"attributes_shots_current", 
                                    "environment.children":"environment_children", 
                                    "environment.dogs":"environment_dogs", 
                                    "environment.cats":"environment_cats",
                                    "contact.email":"contact_email", 
                                    "contact.phone": "contact_phone",
                                    "contact.address.address1": "contact_address_address1", 
                                    "contact.address.city":"contact_address_city", 
                                    "contact.address.state":"contact_address_state", 
                                    "contact.address.postcode":"contact_address_postcode"})
print(new_dog_api_df.head())
```

```python
new_new_dog_api_df = new_dog_api_df.drop(['dog_id'], axis=1)
print(new_new_dog_api_df.head())
```

```python
new_new_dog_api_df.to_sql(name="pet_finder", con=engine, if_exists='append',index = False)
```
