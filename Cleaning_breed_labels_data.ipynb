```python
import os
import pandas as pd
```

```python
breed_labels_path = os.path.join("Resources","original","BreedLabels.csv")

df_breed_labels = pd.read_csv(breed_labels_path) 
df_breed_labels
```

```python
#Drop type = 2 which are Cat breeds
df_breed_labels= df_breed_labels[df_breed_labels['Type']==1]
#Rename columns
df_breed_labels= df_breed_labels.rename(columns = {'BreedID': 'breed_ID', 'Type': 'animal_type', 'BreedName': 'breed_name'})

df_breed_labels
```

```python
#Output pandas dataframe into CSV file and store it in 'clean' folder under Resources folder
output_path = os.path.join("Resources","clean","df_breed_labels_clean.csv")
print(output_path)
df_breed_labels.to_csv(output_path)
```
